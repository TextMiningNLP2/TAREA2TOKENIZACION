```python
#Importar librerías
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist
from sklearn.feature_extraction.text import CountVectorizer
import matplotlib.pyplot as plt
import nltk
import re
#Descargar punkt para poder tokenizar
nltk.download('punkt')

def analyze_tokens(url):
    #Abrimos el archivo, si desea subir otro archivo cambie el path
    # ==============================================================================
    file = open(url,'r')  
    raw_file = file.read() 
    
    #Limpieza del texto
    # ==============================================================================
    #Convertir a minusculas
    nuevo_texto = raw_file.lower()
    #Eliminar protocolo http/https
    nuevo_texto = re.sub('http\S+', ' ', nuevo_texto)
    #Eliminar caracteres especiales
    regex = '[\\!\\"\\#\\$\\%\\&\\\'\\(\\)\\*\\+\\,\\-\\.\\/\\:\\;\\<\\=\\>\\?\\@\\[\\\\\\]\\^_\\`\\{\\|\\}\\~]'
    nuevo_texto = re.sub(regex , ' ', nuevo_texto)
    #Eliminar numeros
    nuevo_texto = re.sub("\d+", ' ', nuevo_texto)
    #Eliminar espacios en blanco múltiples
    nuevo_texto = re.sub("\\s+", ' ', nuevo_texto)
    
    #tokenizar del archivo
    # ==============================================================================
    tokens = word_tokenize(nuevo_texto)
    
    #Crear un objeto de tipo Vectorizer 
    # ==============================================================================
    vectorizer = CountVectorizer()
    vectorizer.fit(tokens)
    
    #Distribución del vocabulario
    # ==============================================================================
    fdist = FreqDist(tokens)
    
    #Imprimir la cantidad de tokens del archivo
    # ==============================================================================
    print("Cantidad de tokens en el archivo: " + str(len(tokens)) + "\n")
    
    #Imprimir el vocabulario y la cantidad por token único
    # ==============================================================================
    print("Tamaño del vocabulario: " + str(len(vectorizer.vocabulary_)))
    print("Vocabulario: ", vectorizer.vocabulary_)    
    
    #Graficar la distribución de los mas comunes 
    # ==============================================================================
    fdist.plot(30, title='Distribución de frecuencia de los 30 tokens más comunes en nuestra colección de texto')
    
```

```python
def multiple_tokens(theList):
    token = []
    vocabulary = []
    
    for element in theList:
        file = open(element,'r')  
        raw_file = file.read()
        nuevo_texto = raw_file.lower()        
        nuevo_texto = re.sub('http\S+', ' ', nuevo_texto)    
        regex = '[\\!\\"\\#\\$\\%\\&\\\'\\(\\)\\*\\+\\,\\-\\.\\/\\:\\;\\<\\=\\>\\?\\@\\[\\\\\\]\\^_\\`\\{\\|\\}\\~]'
        nuevo_texto = re.sub(regex , ' ', nuevo_texto)    
        nuevo_texto = re.sub("\d+", ' ', nuevo_texto)    
        nuevo_texto = re.sub("\\s+", ' ', nuevo_texto)    
        tokens = word_tokenize(nuevo_texto)
        token.append(len(tokens))
        #print(len(tokens))
        vectorizer = CountVectorizer()
        vectorizer.fit(tokens)
        vocabulary.append(len(vectorizer.vocabulary_))
        #print(len(vectorizer.vocabulary_))
    
    x = token
    y = vocabulary
    plt.xlabel("Tokens")
    plt.ylabel("Vocabulary")
    plt.plot(x,y)
    plt.show()
```
